```julia
function overlay_subject_data(fig, subject_data_file, ms=:circle, color=:red, size=10)
    """
    Overlays subject data points on the existing surface plots.
    
    Parameters:
    -----------
    fig : GLMakie.Figure
        The figure with surfaces created by plot_vt_surfaces
    subject_data_file : String
        Path to the MAT file containing subject data
    ms : Symbol
        Marker symbol to use for data points
    color : Symbol or RGB
        Color for the subject data points
    size : Number
        Size of the markers
        
    Returns:
    --------
    fig : Figure
        The updated figure with overlaid data points
    """
    
    # Load subject data
    data = matread(subject_data_file)
    
    # Extract relevant fields - adjust these based on your actual data structure
    subject_distances = data["distance"]  # Assuming this exists
    subject_durations = data["duration"]  # Assuming this exists
    subject_peak_speeds = data["peak_speed"]  # Assuming this exists
    #ct is time_valuation
    
    # Convert time valuations if needed (same as in plot_vt_surfaces)
    
    # Get the existing axes from the figure
    axes = fig.content
    duration_axis = axes[1]
    peak_speed_axis = axes[3]  # Assuming this is the second 3D axis
    
    # Add data points to the duration surface
    GLMakie.scatter!(duration_axis, 
                    subject_ct_mech, 
                    subject_distances, 
                    subject_durations,
                    markersize=size, 
                    color=color, 
                    marker=ms)
    
    # Add data points to the peak speed surface
    GLMakie.scatter!(peak_speed_axis, 
                    subject_ct_mech, 
                    subject_distances, 
                    subject_peak_speeds,
                    markersize=size, 
                    color=color, 
                    marker=ms)
    
    return fig
end
```
